```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
def getImgUrl(url): #ChennelURL
    rt = []
    res = requests.get(url)
    for i in BeautifulSoup(res.text,'html.parser').find_all('img'):
        if 'i.ytimg.com' in i['src']:
            d = i['src'].find('?')
            rt.append(i['src'][:d])
    return rt
```

```python
def downloadImg(urls): #imgURLs
    for i in urls:
        file_name = str(time.time()).replace('.','')
        with open('./p/{}.jpg'.format(file_name) , 'wb') as f:
            res = requests.get(i)
            f.write(res.content)
        print(file_name,' OK!')
```

```python
def saveImg(urls): #ChennelUrls
    for url in urls:
        with open('urls_p.txt','a+') as f:
            f.write(url + '\n')
        imgUrls = getImgUrl(url)
        downloadImg(imgUrls)
```

```python
urls = [
'https://www.youtube.com/user/siemreapfamily/videos',
'https://www.youtube.com/channel/UCBI3JuN_hyyNxi4fKLtISfg/videos',
'https://www.youtube.com/channel/UCZ6mXqoumW8pk7CbYJX0oUA/videos',
'https://www.youtube.com/channel/UCBi47OwQg0pTWYlMfTcA8fA/videos',
'https://www.youtube.com/channel/UCC-7t96Wrq1d5lwygxOvufQ/videos',
'https://www.youtube.com/channel/UCUMm-lEYEyc5BvfgpTr90Mw/videos',
'https://www.youtube.com/channel/UCllZhC5I9J1lffFMxnN3pDg/videos'
]
```
